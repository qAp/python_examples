## Order of plot has effect on shared axes' limits

```python
%matplotlib inline
import os
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
```

```python
matplotlib.rc('font', size=13)
```

```python
x = np.linspace(-100, 100, 1000)
```

```python
# the order of plotting y1 and y2 does not change y-axis limitis
y1 = 5e0 * np.sin(x / 5) + 7
y2 = 2e0 * np.sin(x / 5) + 7
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))
    
axs[0].plot(x, y1)
axs[1].plot(x, y2)
    

axs[0].get_shared_y_axes().join(*axs)
all([ax.set_yscale('log') == None for ax in axs])
```

```python
y1 = 5e0 * np.sin(x / 5) + 7
y2 = 2e0 * np.sin(x / 5) + 7


fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))
    
axs[0].plot(x, y2)
axs[1].plot(x, y1)
    

axs[0].get_shared_y_axes().join(*axs)
all([ax.set_yscale('log') == None for ax in axs])
```

```python
# the order of plotting y1 and y2 changes y-axis limitis
y1 = 5e5 * np.sin(x / 5) + 5.5e5
y2 = 2e5 * np.sin(x / 5) + 5.5e5
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))
    
axs[0].plot(x, y1)
axs[1].plot(x, y2)
    

axs[0].get_shared_y_axes().join(*axs)
all([ax.set_yscale('log') == None for ax in axs])
```

```python
y1 = 5e5 * np.sin(x / 5) + 5.5e5
y2 = 2e5 * np.sin(x / 5) + 5.5e5


fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))
    
axs[0].plot(x, y2)
axs[1].plot(x, y1)
    

axs[0].get_shared_y_axes().join(*axs)
all([ax.set_yscale('log') == None for ax in axs])
```
