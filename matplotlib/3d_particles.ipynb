```python
%matplotlib notebook 

import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
import pandas as pd
```

```python
coords = pd.DataFrame(columns=['x', 'y', 'z'])
```

```python
coords.loc['A'] = [0, 0, 0]
coords.loc['B'] = [12, 15, 18]
coords.loc['C'] = [25, 35, 28]
coords.loc['F'] = [21, 12, 29]
coords.loc['E'] = [21, 14, 15]
coords.loc['J'] = [22, 32, 27]
```

```python
coords
```

```python
x = coords['x']
y = coords['y']
z = coords['z']

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(x, y, z, c=z, cmap='Purples')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
